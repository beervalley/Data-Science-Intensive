# Question 1: which ten company get the most fund in total?

```python
table = data[['company_name', 'raised_amount']]
table.head(10)
```

```python
table.tail(10)
```

```python
table.groupby('company_name').sum().sort('raised_amount', ascending=False)
```

```python
table.groupby('company_name').sum().sort('raised_amount', ascending=False).head(10)
```

### Above are the ten companies that raised the most fund in all rounds.

# Question 2: is there a trend that the raised fund increases along with the year?

```python
trend = data[['company_name', 'raised_amount', 'funded_year']]
trend.head(10)
```

```python
trend.tail(10)
```

```python
trend_ = trend.groupby('funded_year').sum().sort(ascending=False).head(29)
trend_
```

```python
trend_.plot()
```

### Answer for question 2: yes or no from the above figure. 

### From 2004 unitl now,  we see that the total fund amound increases signifcantly along with the year.
### From 1974 to 2003, we see that the total fund amount is almost flat and only increase from 1999 to 2000, and then from 2000 to 2004 there is slight decrease and increase alternatively.
