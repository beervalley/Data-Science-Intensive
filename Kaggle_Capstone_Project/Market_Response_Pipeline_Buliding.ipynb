### 3. Concatenate the original numerical DataFrame and the generated dummy DataFrame

```python
# concatenate the original DataFrame and the dummy DataFrame (axis=0 means rows, axis=1 means columns)
data_ = pd.concat([data_num, data_9], axis=1)
data_.sample(n=5, random_state=1)
```

### 4. Filling the Missing Data with Mean, or 0.

```python
data_ = data_.fillna(data_num.mean())
data_ = data_.fillna(0)
data_.head()
```

```python
data_array_origin = data_.as_matrix()
data_array_origin
```

### 5. Feature Scaling and Normalization for the Dataset

```python
# do the feature scaling and normalization for the data set
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
data_scaled = min_max_scaler.fit_transform(data_array_origin)
print data_scaled
```

```python
from sklearn.preprocessing import normalize

data_normal = normalize(data_scaled, norm='l2')
data_normal
```

### 6. Feature Binarization for the normalized data

```python
from sklearn.preprocessing import Binarizer

binarizer = Binarizer().fit(data_normal)
data_binary = binarizer.transform(data_normal)
```

### 7. Feature Selection based on Variance

```python
# feature selection based on variance

from sklearn.feature_selection import VarianceThreshold

sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
data_var = sel.fit_transform(data_binary)
data_var
```

```python
data_var.shape
```

### 8. Tree-based Feature Selection

```python
# Tree-based feature selection
from sklearn.ensemble import ExtraTreesClassifier

clf = ExtraTreesClassifier()
clf.fit(data_var, target)
```

```python
feature = clf.transform(data_var)
feature.shape
```

### 9. Cross Validation, Model Selection, Grid Search for Parameter Tuning, Model Training and Prediction

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.grid_search import GridSearchCV

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000] }

clf = GridSearchCV(LogisticRegression(penalty='l2'), param_grid)
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
```

### 10. Evaluation Metrics with ROC Curve and AUC Score

```python
from sklearn.metrics import accuracy_score
print 'Accuracy of the Logistic Regression Model:', accuracy_score(y_test, pred)

from sklearn.metrics import f1_score
print 'f1 score:', f1_score(y_test, pred)
```

```python
from ggplot import *
from sklearn import metrics

fpr, tpr, _ = metrics.roc_curve(y_test, pred)

df = pd.DataFrame(dict(fpr=fpr, tpr=tpr))
ggplot(df, aes(x='fpr', y='tpr')) +\
    geom_line() +\
    geom_abline(linetype='dashed')
```

```python
auc = metrics.auc(fpr,tpr)
ggplot(df, aes(x='fpr', ymin=0, ymax='tpr')) +\
    geom_area(alpha=0.2) +\
    geom_line(aes(y='tpr')) +\
    ggtitle("ROC Curve w/ AUC=%s" % str(auc))
```
