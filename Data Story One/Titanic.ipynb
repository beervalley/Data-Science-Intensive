```python
age_sur_fraction = [[0, 62, 38], [10, 102, 41], [20, 397, 129], [30, 167, 73], [40, 89, 34], [50, 48, 20], [60, 19, 6], [70, 6, 0], [80, 1, 1]]
```

```python
age_sur_fraction = pd.DataFrame(age_sur_fraction)
age_sur_fraction.columns = ['age_period', 'number_of_people', 'number_of_survival']
age_sur_fraction
```

```python
age_sur_fraction['survival_fraction'] =  age_sur_fraction.number_of_survival / age_sur_fraction.number_of_people
```

```python
age_sur_fraction
```

```python
#Can we find any correlation between the age period and the probability of survival? 
age_sur_fraction.plot(x='age_period', y='survival_fraction', kind='bar')
```

```python
# From the plot, we can see that children have a high probability to servive from the disaster.
```
